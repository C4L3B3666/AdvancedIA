```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

print("Ambiente está pronto!")
```

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 4, 6, 8, 10])

modelo = LinearRegression()
modelo.fit(X, y)

nova_entrada = np.array([[6]])
predicao = modelo.predict(nova_entrada)

print(f"Entrada: {nova_entrada.flatten()[0]} → Previsão: {predicao.flatten()[0]}")
```

```python
plt.scatter(X, y, color='blue', label='Dados')
plt.plot(X, modelo.predict(X), color='red', label='Regressão Linear')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Teste de Regressão Linear")
plt.legend()
plt.grid(True)
plt.show()
```
